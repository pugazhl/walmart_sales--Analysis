```python

df = pd.read_sql(
    """
    SELECT 
        branch,
        DAYNAME(STR_TO_DATE(date, '%%m/%%d/%%y')) AS day_name,
        COUNT(*) AS no_transactions 
    FROM walmart 
    GROUP BY branch, day_name
    """, 
    con=engine_mysql
)
df
```

```python
df = pd.read_sql('SELECT category,SUM(total) as total_revenue, SUM(total * profit_margin) as profit FROM walmart group by 1', con=engine_mysql)
df
```

```python
# df = pd.read_sql('SELECT Branch, payment_method, COUNT(*) as total_trans,RANK() OVER(PARTITION BY Branch ORDER BY COUNT(*) DESC) as rank FROM walmart', con=engine_mysql)
# df
df = pd.read_sql(
    """
    SELECT 
        category,
        SUM(unit_price * quantity) AS total_sales,
        SUM(unit_price * quantity * profit_margin) AS total_profit
    FROM walmart
    GROUP BY category
    ORDER BY total_sales DESC
    """,
    con=engine_mysql
)
df
```

```python
df = pd.read_sql(
    """
    SELECT 
        City,
        category,
        MIN(rating) as min_rating,
        MAX(rating) as max_rating,
        AVG(rating) as avg_rating
    FROM walmart
    GROUP BY City, category
    ORDER BY City, category
    """, 
    con=engine_mysql
)
df
```

```python
df = pd.read_sql(
    """
    SELECT 
        CASE 
            WHEN HOUR(time) < 12 THEN 'morning'
            WHEN HOUR(time) BETWEEN 12 AND 17 THEN 'afternoon'
            ELSE 'evening'
        END AS day_time,
        COUNT(*) AS transaction_count
    FROM walmart
    GROUP BY day_time
    ORDER BY 
        CASE day_time
            WHEN 'morning' THEN 1
            WHEN 'afternoon' THEN 2
            ELSE 3
        END
    """,
    con=engine_mysql
)
df
```
